```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
cars = sm.datasets.get_rdataset('mtcars').data
print(cars.head(4).T)
print(cars.info())
```

```python
cars
```

```python
sns.regplot(x=cars.mpg,y=cars.wt,ci=95)
corcoef = np.corrcoef(cars.mpg,cars.wt)[0,1]
text_cc = '\n\n\n\ncorrelation coefficient is %1.4f'%corcoef
plt.title('Scatter plot MPG vs Weight')
plt.suptitle(text_cc)
plt.gcf().tight_layout()

plt.show()
```

```python
sns.heatmap(cars.corr(),cmap='BuGn',annot=True,fmt='.2f')
plt.show()
```

```python
X = cars.wt
y = cars.mpg

X = sm.add_constant(X)
model = sm.OLS(y,X).fit()
predictions = model.predict(X)

model.summary()
```

```python
cars['predictions'] = model.predict(X)
residuals = model.resid
ax = sns.residplot(cars.predictions,residuals,lowess = True,color='g')
ax.set(xlabel = 'Fitted value',ylabel = 'Residuals',title = 'Residuals vs ftted vlues plot\n')
plt.show()
```

```python
sns.distplot(residuals);
```

```python
SSE = 0
SST = 0

for i in range(len(predictions)):
  SSE += (predictions[i]-y[i])**2

ymean = np.mean(y)

for i in range(len(predictions)):
  SST += (y[i] - ymean)**2
  
print('SSE = %5.4f SST = %5.4f' %(SSE,SST))

Rsq = np.round(1-(SSE/SST),4)

print('Calculated %s = %1.3f'%('R Squared',Rsq))
```

```python
def calculate_vif(x):
    thresh = 5.0
    output = pd.DataFrame()
    k = x.shape[1]
    vif = [variance_inflation_factor(x.values, j) for j in range(x.shape[1])]
    for i in range(1,k):
        print("Iteration no.")
        print(i)
        print(vif)
        a = np.argmax(vif)
        print("Max VIF is for variable no.:")
        print(a)
        if vif[a] <= thresh :
            break
        if i == 1 :          
            output = x.drop(x.columns[a], axis = 1)
            vif = [variance_inflation_factor(output.values, j) for j in range(output.shape[1])]
        elif i > 1 :
            output = output.drop(output.columns[a],axis = 1)
            vif = [variance_inflation_factor(output.values, j) for j in range(output.shape[1])]
    return(output)
```
